**Classes**
1. used to organize code 
    * combining data(attribute) and actions(method)
    * a blue print for creating objects
2. to use a class, create an *instance*
3. **attribute** - a variable associated with a class, the data
4. **method** - a function associated with a class, functionality
5. * both attributes and methods can be accessed via dot notation
    * ex 'Fido.tail.wags()'
6. start class names with a Capital Letter
    * for ex.  'class ClassName:'
    * or, 'class Dog:'
7. all classes hava a function called __init__()
    * it's always executed when the class is being initiated
    * the __init__() method runs automatically
    * __init__() requires one argument and one optional parameter
    * __init__.py might also be used.
8. when creating class, don't bother with the parentheses after the class name, unless you are subclassing other classes, 
    * which would look like 'class TestClass(subclass):'
9. **self** 
    * first parameter, 'self', references the current instance of the class 
    * required in the method definition, must come first 
    * any variable prefixed with with 'self' is available to every method in the class
    * is a convention, not a real Python keyword
    * when Python calls the __init__ method to create an instance of the class, the method will automatically pass the self argument.

```python
# now let's create my own, how about bikes?
class Bike:
    """this class will represent bicycles"""
    # the parameters are wheelsize, discipline, model, etc
    def __init__(self, wheelSize, discipline, make, model):
        self.wheelSize = wheelSize
        self.discipline = discipline
        self.make = make
        self.model = model
        
    def bike_details(self):
        print("My main " + self.discipline + " bike is a " + self.make + " " + self.model +".") 
```

```python
# parens = tuples

bike1 = Bike(700, "gravel", "Trek", "Checkpoint")
bike2 = Bike(700, "do it all", "Bianchi", "Volpe")
bike3 = Bike(26, "urban assault", "Trek", "Antelope")
bike4 = Bike(26, "full squish", "Specialized", "SX Trail II")
bike5 = Bike(26, "hard tail", "KHS", "Alite 1000")
```

```python
print(bike1.wheelSize)
print(bike1.discipline)
print(bike1.make)
print(bike1.model)
```

```python
# Let's get the details for bike 1
bike1.bike_details()
bike2.bike_details()
bike3.bike_details()
bike4.bike_details()
bike5.bike_details()
```

```python
# now let's use a contstructor to help with attribute errors, 
# like if your object attribute and is mispelled in the attribute vs function
class Robot:
    def __init__(self, name, color, weight):
        # this block of code is setting set this object's name to givenName, etc
        # it's common practice to name your arguments above the same as your attributes below
        self.name = name
        self.color = color 
        self.weight = weight
        
    def introduce_self(self):
        print("My name is " + self.name)
     
```

```python
r1 = Robot("Tom", "red", 30)
r2 = Robot("Jerry", "blue", 40)
```

```python
r1.introduce_self()
r2.introduce_self()
```

```python
# another class example, this time from Python Crash Course book, page 162
# Github  = https://github.com/ehmatthes/pcc
class Dog:
    """a simple attempt to model a dog"""
    
    def __init__(self, name, age):
        """inititalize name and age attributes"""
        self.name = name
        self.age = age
        
    def sit(self):
        """simulate a dog sitting in response to a command."""
        print(self.name.title() + " is now sitting.")
        
    def roll_over(self):
        """simulate a dog rolling over"""
        print(self.name.title() + " rolled over!")
        
```

```python
class Dog:
    """A simple attempt to model a dog."""
    
    def __init__(self, name, age):
        """Initialize name and age attributes."""
        self.name = name
        self.age = age
        
    def sit(self):
        """Simulate a dog sitting in response to a command."""
        print(self.name.title() + " is now sitting.")
        
    def roll_over(self):
        """Simulate rolling over in response to a command."""
        print(self.name.title() + " rolled over!")
        
    #creating the buddy bit from DataCamp article
    def setBuddy(self, buddy):
        self.buddy = buddy
```

```python
my_dog = Dog('willie', 6)
```
